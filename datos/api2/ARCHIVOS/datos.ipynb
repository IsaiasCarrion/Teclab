```python
# Importamos de vuelta el archivo pero sin las ultimas 3 filas
macro = pd.read_csv("datos_macro.csv", skipfooter = 3, engine = "python")

# Filtramos las observaciones/filas que correspondan a los paises y variables que queremos
macro_f = macro [(macro["Country"].isin(["Argentina","Chile","Brazil"])) &
                 (macro["Subject Descriptor"] == "Gross domestic product, constant prices") &
                 (macro["Units"] == "Percent change")]

macro_f
```

```python
# Eliminamos las columnas que ya no necesitamos
macro_f = (macro_f.drop(columns = ["WEO Subject Code", "ISO", "WEO Country Code",
                        "Subject Notes", "Subject Descriptor","Units", "Scale",
                        "Country/Series-specific Notes", "Estimates Start After"]))
macro_f
```

```python
# Transportamos la tabla para tenerla en formato largo, dejando fijo el pais,
# Convirtiendo los aÃ±os de columnas a filas, y definiendo la variable como crecimiento_pbi
macro_tr = pd.melt(macro_f, id_vars= ["Country"], var_name= "year", value_name = "crecimiento_pbi")
macro_tr
```

```python
# Renombramos la palabra country por pais
macro_tr = macro_tr.rename(columns = {"Country":"pais"})
macro_tr
```

```python
# Ordenamos las filas por pais y por aÃ±o
macro_tr = macro_tr.sort_values(by=["pais", "year"])
macro_tr
```

```python
# Tipos de variables
macro_tr.dtypes
```

```python
tc_br.dtypes
```

```python
# Convertimos algunos de los campos a tipos de variables que nos sirvan: aÃ±os en enteros, crecimiento como real
macro_tr = macro_tr.astype({"pais": str, "year": int, "crecimiento_pbi": float})
macro_tr.dtypes
```

```python
# Filtramos observaciones anteriores a 2020, el ultimo aÃ±o con datos completos
macro_tr = macro_tr[macro_tr["year"] <= 2020]
macro_tr
```

```python
# Calculamos las principales estadisticas descriptivas de la variable crecimiento
macro_tr["crecimiento_pbi"].describe()
```

```python
# Concatenamos las tablas con el tipo de cambio de Brasil y de Argentina
tc = pd.concat([tc_br, tc_ar])

# Combinamos las dos tablas: la de tiene la serie del crecimiento del PBI y la de tipo de cambio
macro_tc = pd.merge(macro_tr, tc, left_on = ["pais", "year"], right_on = ["pais", "year"], how = "outer")
macro_tc
```

```python
# Agregamos por pais, usando el promedio (promedio de la tasa de crecimiento del PBI)
macro_ag = macro_tr.groupby(["pais"], as_index = False).agg({"crecimiento_pbi":"mean"})
macro_ag
```
