## Base de datos.

En esta base de datos ficticia, cada fila representa una transacción de compra realizada por un cliente. La tabla incluye las siguientes columnas:

##### ID Cliente: Un identificador único para cada cliente.

##### Fecha de Compra: La fecha en que se realizó la compra.

##### Monto de la Compra: El monto total de la compra realizada por el cliente.

##### Segmento de Cliente: Indica si el cliente es nuevo o recurrente.

##### Satisfacción del Cliente: Una calificación (del 1 al 5) que indica el nivel de satisfacción del cliente con la compra.


```python
###################################### CODIGO PARA GENERAR BASE DE DATOS DE PRUEBA #######################################
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Definir el número de clientes y el rango de fechas
num_clientes = 1000
fecha_inicio = datetime(2022, 1, 1)
fecha_fin = datetime(2024, 12, 31)

# Generar datos aleatorios para la base de datos
ids_clientes = np.arange(1, num_clientes + 1)
fechas_compra = [np.random.choice(pd.date_range(start=fecha_inicio, end=fecha_fin), size=np.random.randint(1, 5)) for _ in range(num_clientes)]
montos_compra = [np.random.randint(10, 200, size=len(fechas)) for fechas in fechas_compra]
segmentos = np.random.choice(['Nuevo', 'Recurrente'], size=num_clientes)
satisfaccion_cliente = np.random.randint(1, 6, size=num_clientes)

# Crear una lista de todas las transacciones
transacciones = []
for i in range(num_clientes):
    for j in range(len(fechas_compra[i])):
        transacciones.append([ids_clientes[i], fechas_compra[i][j], montos_compra[i][j], segmentos[i], satisfaccion_cliente[i]])

# Crear un DataFrame a partir de la lista de transacciones
df = pd.DataFrame(transacciones, columns=['ID Cliente', 'Fecha de Compra', 'Monto de la Compra', 'Segmento de Cliente', 'Satisfacción del Cliente'])

# Guardar el DataFrame como un archivo CSV
df.to_csv('base_datos_ventas.csv', index=False)

# Mostrar las primeras filas del DataFrame
df
```

## Hipótesis 2: Diferencia entre Nuevos y Clientes Recurrentes:

```python
import pandas as pd

# Cargar los datos desde el archivo CSV
df = pd.read_csv('base_datos_ventas.csv')

# Segmentación de Clientes
ventas_nuevos = df[df['Segmento de Cliente'] == 'Nuevo']['Monto de la Compra'].sum()
ventas_recurrentes = df[df['Segmento de Cliente'] == 'Recurrente']['Monto de la Compra'].sum()

# Análisis de Retención de Clientes
clientes_nuevos = df[df['Segmento de Cliente'] == 'Nuevo']['ID Cliente'].nunique()
clientes_recurrentes = df[df['Segmento de Cliente'] == 'Recurrente']['ID Cliente'].nunique()

compras_recurrentes = df[df['Segmento de Cliente'] == 'Recurrente'].groupby('ID Cliente')['Fecha de Compra'].count().sum()
compras_nuevos = df[df['Segmento de Cliente'] == 'Nuevo'].groupby('ID Cliente')['Fecha de Compra'].count().sum()

proporcion_compras_recurrentes = compras_recurrentes / clientes_recurrentes
proporcion_compras_nuevos = compras_nuevos / clientes_nuevos

# Comparación de las ventas totales generadas por cada grupo
print("Ventas totales de clientes nuevos:", ventas_nuevos)
print("Ventas totales de clientes recurrentes:", ventas_recurrentes)

# Comparación de la proporción de compras repetidas entre los dos grupos
print("Proporción de compras repetidas de clientes recurrentes:", proporcion_compras_recurrentes)
print("Proporción de compras repetidas de clientes nuevos:", proporcion_compras_nuevos)
```

### Hipótesis 2:
Podría haber una diferencia significativa en las ventas entre nuevos clientes y clientes recurrentes, siendo los nuevos clientes menos propensos a realizar compras repetidas.

### Conclusión para la Hipótesis 2:
Las ventas totales generadas por clientes nuevos son de $127.818,

mientras que las ventas totales generadas por clientes recurrentes son de $124.849. Esto indica que, durante el período en cuestión, los clientes nuevos generaron más ventas en comparación con los clientes recurrentes.

La proporción de compras repetidas de clientes recurrentes es de aproximadamente 2.40, mientras que la proporción de compras repetidas de clientes nuevos es de aproximadamente 2.46. Esto sugiere que los clientes nuevos tienen una ligera mayor propensión a realizar compras repetidas en comparación con los clientes recurrentes.

### Implicaciones:
Contrario a lo esperado por la hipótesis, los nuevos clientes no son menos propensos a realizar compras repetidas en comparación con los clientes recurrentes. De hecho, la proporción de compras repetidas es ligeramente mayor entre los nuevos clientes.

Aunque los clientes recurrentes pueden tener una menor proporción de compras repetidas, esto no parece haber afectado significativamente sus contribuciones a las ventas totales, ya que las ventas totales generadas por clientes nuevos son mayores que las de clientes recurrentes durante el período analizado.

## Hipótesis 3: Patrón Estacional en las Ventas

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Cargar los datos desde el archivo CSV
df = pd.read_csv('base_datos_ventas.csv')

# Convertir la columna 'Fecha de Compra' a tipo datetime y establecerla como índice
df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'])
df.set_index('Fecha de Compra', inplace=True) # establece la columna 'Fecha de Compra' como el índice del DataFrame

# Agrupar las ventas por mes y calcular las ventas totales
# ealiza una operación de re-muestreo en los datos temporales del 
# DataFrame df para agrupar los datos por mes ('M') y calcular la suma de las ventas ('Monto de la Compra') para cada mes.
ventas_por_mes = df.resample('M')['Monto de la Compra'].sum()

# Graficar las ventas totales en función del tiempo (por mes)
print("Graficar las ventas totales en función del tiempo por mes")
print('El gráfico de línea que muestra las ventas totales en función del tiempo (por mes) para identificar patrones estacionales o tendencias a lo largo del año.')
plt.figure(figsize=(10, 6))
ventas_por_mes.plot()
plt.title('Ventas Totales por Mes')
plt.xlabel('Fecha')
plt.ylabel('Ventas Totales')
plt.grid(True)
plt.show()

# Descomposición de Series Temporales
# descomponer una serie temporal en sus componentes de tendencia, estacionalidad y residuos.
decomposicion = seasonal_decompose(ventas_por_mes, model='additive')

# Graficar la descomposición
print("Series Temporales de ventas por mes")
print('El conjunto de subgráficos que muestran la descomposición de la serie temporal en componentes de tendencia, estacionalidad y residuos.')

plt.figure(figsize=(10, 8))

plt.subplot(411)
plt.plot(decomposicion.observed, label='Observado')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(decomposicion.trend, label='Tendencia')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(decomposicion.seasonal, label='Estacionalidad')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(decomposicion.resid, label='Residuos')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np

# Ajustar el modelo ARIMA
modelo = ARIMA(ventas_por_mes, order=(1, 1, 1))  # Aquí se especifican los parámetros p, d y q del modelo ARIMA
ajuste = modelo.fit()

# Hacer predicciones para los próximos 12 meses
predicciones = ajuste.forecast(steps=12)  # Se especifica el número de pasos hacia adelante para predecir

# Imprimir las predicciones
print("Predicciones para los próximos 12 meses:")
print(predicciones)
```

### Hipótesis 3:
Existe la posibilidad de que haya un patrón estacional en las ventas, con una disminución en las ventas durante ciertos meses del año.


### Conclusión para la Hipótesis 3
Basándonos en los gráficos de tendencias temporales y la descomposición de series temporales, podemos obtener algunas conclusiones:

#### Análisis de Tendencias Temporales:

Observamos una variabilidad en las ventas a lo largo del tiempo. No parece haber una tendencia clara de aumento o disminución en las ventas durante el período analizado.

Sin embargo, parece haber cierta estabilidad en las ventas, con fluctuaciones regulares en lugar de cambios drásticos.

#### Descomposición de Series Temporales:

La descomposición de la serie temporal muestra componentes de tendencia, estacionalidad y residuos.

La componente de tendencia parece ser relativamente estable, con algunas fluctuaciones a lo largo del tiempo, pero sin una tendencia clara de aumento o disminución.

La componente de estacionalidad muestra patrones estacionales en las ventas. Podemos observar aumentos y disminuciones en las ventas que se repiten a lo largo del período analizado.

Los residuos representan las variaciones aleatorias en las ventas que no pueden ser explicadas por la tendencia o la estacionalidad.

#### En resumen, según el análisis realizado, podemos concluir que:

Existe evidencia de patrones estacionales en las ventas, con aumentos y disminuciones que se repiten a lo largo del año.

No parece haber una tendencia clara de aumento o disminución en las ventas durante el período analizado.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Cargar los datos desde el archivo CSV
df = pd.read_csv('base_datos_ventas.csv')

# Convertir la columna 'Fecha de Compra' a tipo datetime y establecerla como índice
df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'])


# Análisis de Opiniones y Comentarios de Clientes
plt.figure(figsize=(10, 6))
plt.hist(df['Satisfacción del Cliente'], bins=5, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histograma de Calificaciones de Satisfacción del Cliente')
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()


# Convertir la fecha de compra a mes y año
df['Año-Mes'] = df['Fecha de Compra'].dt.to_period('M')

# Calcular el promedio de calificaciones por mes
promedio_calificaciones_mes = df.groupby('Año-Mes')['Satisfacción del Cliente'].mean()

# Graficar el promedio de calificaciones por mes
plt.figure(figsize=(10, 6))
promedio_calificaciones_mes.plot(marker='o', color='b')
plt.title('Promedio de Calificaciones de Satisfacción del Cliente por Mes')
plt.xlabel('Mes')
plt.ylabel('Promedio de Calificaciones')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
